```python
import numpy as np

from math import pow

from math import sqrt

def mydeviation(myarray,kind):
    
    resultarray=[]
    
    squaresarray=[]
    
    average=np.mean(myarray)
    
    
    for i in myarray:
        
        
        resultarray.append(i-average)
        
        
    #print(resultarray)
    
    
    for j in resultarray:
        
        squaresarray.append(pow(j,2))
        
    #print(squaresarray)
    
    noarray=np.array(squaresarray)
    #print(noarray)
    
    if kind=="population":
        
        squarestandarddevation=np.mean(noarray)
        
    elif kind=="sample":
        
        squarestandarddevation=np.sum(noarray)/(len(noarray)-1)
        
    else:
        
        print("kind should be population or sample")
        
  
    #print(meanarray)
    
    print(kind + "  standarad deviation is ")
    
    print(sqrt(squarestandarddevation))
    
mydeviation([1550,1700,900,850,1000,950],"population")

print(" ")

mydeviation([1550,1700,900,850,1000,950],"sample")
    
    
```
